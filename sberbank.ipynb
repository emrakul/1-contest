```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
data = pd.read_csv("transactions.csv")
types = pd.read_csv("tr_types.csv")
gender = pd.read_csv("customers_gender_train.csv")
```

```python
#codes = pd.read_csv("tr_mcc_codes.csv")
transactions = data
```

```python
transactions.tr_datetime = data.tr_datetime.str.split()
```

```python
transactions.tr_datetime = [int(i[1][:2]) for i in transactions.tr_datetime]
```

```python
gender_transactions = pd.merge(gender, transactions, on='customer_id' )
```

```python
gender_transactions.drop(axis=1, labels='term_id', inplace=True)
```

```python
gender_transactions.groupby(['customer_id']).aggregate(lambda x: tuple(x))
```

```python
gender_transactions.groupby(['gender']).tr_datetime.hist(bins=40)
```

```python
males, females = [], []
for i in range(len(transactions)):
    if(gender[gender.customer_id == transactions.customer_id.iloc[i]].gender.values[0] == 0):
        males.append(transactions.tr_datetime.iloc[i])
    else:
        females.append(transactions.tr_datetime.iloc[i])

```
